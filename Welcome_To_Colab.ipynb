```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
image = cv2.imread('Wp Couple.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(image)
plt.axis('off')
```

```
resized = cv2.resize(image, None, fx=1.5, fy=1.5)

plt.imshow(resized)
plt.axis('off')
plt.title("Resized")
plt.show()
```

```
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated = cv2.warpAffine(image, matrix, (w, h))

plt.imshow(rotated)
plt.axis('off')
plt.title("Rotated 120")
plt.show()
```

```
rows, cols, _ = image.shape

shear_matrix = np.float32([[1, 0.5, 0],
                           [0, 1, 0]])

sheared = cv2.warpAffine(image, shear_matrix, (cols, rows))

plt.imshow(sheared)
plt.axis('off')
plt.title("Shear")
plt.show()
```

```
negative = 255 - image

plt.imshow(negative)
plt.axis('off')
plt.title("Negative")
plt.show()
```

```
c = 255 / np.log(1 + np.max(image))
log_image = c * (np.log(1 + image))
log_image = np.array(log_image, dtype=np.uint8)

plt.imshow(log_image)
plt.axis('off')
plt.title("Log")
plt.show()
```

```
gamma = 2.0

gamma_img = np.array(255 * (image / 255) ** gamma, dtype='uint8')

plt.imshow(gamma_img)
plt.axis('off')
plt.title("Gamma")
plt.show()
```
